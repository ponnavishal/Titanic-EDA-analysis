```python
sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.xticks([0, 1], ['Died', 'survived'])
plt.show()

df['survived'].value_counts(normalize=True)
```

```python
sns.countplot(x='sex', data=df, hue='survived')
plt.title("survival by Gender")

sns.countplot(x='pclass', data=df, hue='survived')
plt.title("Survival by Passenger Class")

sns.countplot(x='embarked', data=df, hue='survived')
plt.title("Survival by Embarkation")
plt.show()
```

```python
df['age'].hist(bins=30, edgecolor='k')
plt.title("Age Distribution")

df['fare'].hist(bins=30, edgecolor='k')
plt.title("Fare Distribution")
plt.show()
```

```python
sns.boxplot(x='pclass', y='age', data=df)
plt.title("Age by Class")

sns.boxplot(y='fare', data=df)
plt.title("Fare Outliers")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Data Heatmap")
plt.show()
```

```python
sns.pairplot(df[['survived', 'age', 'fare', 'pclass']], hue='survived')
```

```python
plt.figure(figsize=(8, 6))
# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=np.number)
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Only passengers who died
dead = df[df['survived'] == 0]

plt.figure(figsize=(20, 5))
plt.hist(dead['age'].dropna(), bins=30, color='blue', edgecolor='black')
plt.title("Age Distribution of Passengers Who Died")
plt.xlabel("age")
plt.ylabel("Number of Deaths")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['age'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Age Distribution of All Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.grid(True)
plt.show()
```

```python
# Define age groups
df['ageGroup'] = pd.cut(df['age'], bins=[0, 10, 20, 30, 40, 50, 60, 80],
                        labels=['0–10', '11–20', '21–30', '31–40', '41–50', '51–60', '60+'])

# Count passengers in each group
passenger_count = df['ageGroup'].value_counts().sort_index()

# Plot bar chart
passenger_count.plot(kind='bar', color='teal', edgecolor='black')
plt.title("Passenger Count by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Passengers")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
